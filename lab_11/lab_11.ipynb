```python
import pandas as pd
df = pd.read_csv('credits.csv')
df
```

```python
df['transaction_date'].str.split('-',expand =True).iloc[:,1].unique()
```

```python
md = dict(zip(['Oct', 'Aug', 'Apr', 'May', 'Sep', 'Feb', 'Jun', 'Mar', 'Dec',
       'Nov', 'Jul', 'Jan'],
             ['10','8','4','5','9','2','6','3','12','11','7','1'] ))
md
```

```python
df['transaction_date'].apply(lambda x: pd.to_datetime(x.replace(x.split('-')[1],
                                                                md[x.split('-')[1]]),format ='%d-%m-%y'))
```

```python
import sqlite3
con = sqlite3.connect('transactions.db')
df.to_sql('transactions', con=con)
```

```python
%%writefile app.py
import streamlit as st
import matplotlib.pyplot as plt
import numpy as np
import sqlite3
import pandas as pd

con = sqlite3.Connection('transactions.db')
df = pd.read_sql("SELECT * FROM transactions", con)
cities = st.selectbox('выберите город', df.city.unique())
fig,ax =plt.subplots()
data = df.loc[df.city==cities,:]
data.groupby(['gender', 'card_type','exp_type' ])['amount'].agg({'min','mean', 'max'}).plot(kind = 'bar', ax=ax)
st.pyplot(fig)


```
